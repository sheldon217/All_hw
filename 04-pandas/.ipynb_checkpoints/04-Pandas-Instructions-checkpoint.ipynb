```python
whale_returns_csv = pd.read_csv("Starter_Code/Resources/whale_returns.csv")
whale_returns_csv = whale_returns_csv.set_index(pd.to_datetime(whale_returns_csv["Date"])).drop(columns="Date").dropna()
algo_returns_csv = pd.read_csv("Starter_Code/Resources/algo_returns.csv")
algo_returns_csv = algo_returns_csv.set_index(pd.to_datetime(algo_returns_csv["Date"])).drop(columns="Date").dropna()
sp500_history_csv_original= pd.read_csv("Starter_Code/Resources/sp500_history.csv")
sp500_history_csv_copy = sp500_history_csv_original.rename(columns={"Close":"SP500"}).set_index(pd.to_datetime(sp500_history_csv_original["Date"])).drop(columns="Date").dropna().copy()
sp500_history_csv_copy["SP500"] = sp500_history_csv_copy["SP500"].str.replace("$", " ")
sp500_history_csv_copy=pd.to_numeric(sp500_history_csv_copy["SP500"]).to_frame()
sp500_history_csv_copy = sp500_history_csv_copy.pct_change()
return_daily_concat = pd.concat([whale_returns_csv,algo_returns_csv,sp500_history_csv_copy], axis=1, join="inner")
return_daily_concat.rename(columns={"SOROS FUND MANAGEMENT LLC": "Soros", "BERKSHIRE HATHAWAY INC":"Berk", "PAULSON & CO.INC. ":"Paul", "TIGER GLOBAL MANAGEMENT LLC":"Tiger"}, inplace=True)
Tiger = "TIGER GLOBAL MANAGEMENT LLC"
Soros = "SOROS FUND MANAGEMENT LLC"
Paul = "PAULSON & CO.INC. "
Berk = "BERKSHIRE HATHAWAY INC"
Tiger_df = return_daily_concat["Tiger"]
Paul_df = return_daily_concat["Paul"]
Soros_df = return_daily_concat["Soros"]
Berk_df = return_daily_concat["Berk"]
Algo1_df =  return_daily_concat["Algo 1"]
Algo2_df = return_daily_concat["Algo 2"]
SP500 = return_daily_concat["SP500"]
```
